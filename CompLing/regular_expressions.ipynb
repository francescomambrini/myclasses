````python
import re
````

Prima di tutto carichiamo un file di solo testo. Ad es., perché non *La Divina Commedia* di Dante Alighieri? Possiamo scaricare una versione solo testo direttamente dal web, all'indirizzo più sotto

````python
import requests

url = 'https://dmf.unicatt.it/~della/pythoncourse18/commedia.txt'
dante = requests.get(url).text
````

Ispezioniamo i primi 120 caratteri

````python
print(dante[:120])
````

Tra parentesi, così possiamo sapere quanti caratteri ha la Divina Commedia:

````python
len(dante)
````

Vediamo subito che i canti sono preceduti da titoli tipo:

```
Inferno: Canto I
```

Possiamo cercare tutte le righe che cominciano con stringhe simili? Ovviamente sappiamo che le cantiche sono solo 3 (Inferno, Purgatorio, Paradiso), i canti arrivano al massimo a 34 (quindi servono solo i simboli romani: IXV)

Proviamo con un pattern del genere:

```regex
'^Inferno|Purgatorio|Paradiso: Canto [IXV]+'
```

````python
for l in dante.split('\n'):
    if re.search(r'^Inferno|Purgatorio|Paradiso: Canto [IXV]+', l):
        print(l)
````

````python
dante_cleaned = re.sub(r'(Inferno|Purgatorio|Paradiso): Canto [IXV]+', '', dante)
````

Ora la stringa "Canto I" non viene più trovata (mentre viene trovata al carattere 65 nel file originale)

````python
dante.find('Canto I')
````

````python
dante_cleaned.find('Canto I')
````

Possiamo andare avanti e rimpiazzare i caratteri di nuova riga `\n` consecutivi con un solo a capo

````python
dante_cleaned = re.sub(r'(\r\n)+', '\n', dante_cleaned)
````

````python
print(dante_cleaned)
````
